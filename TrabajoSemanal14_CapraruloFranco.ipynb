## 1) Calcular los parámetros S de la siguiente red




$s_{11} $representa el coeficiente de reflexión en el puerto 1 del cuadripolo y $ s_{21} $el coeficiente de transmisión directa.

Analizando el parámetro $s_{11}$: $$ s_{11} = \frac{s^3}{s^3+2s^2+2s+1} $$

Si $ \omega = 0 \Longrightarrow s_{11} = 0 $ (No hay reflexión, el cuadripolo NO filtra bajas frecuencias)

Si $ \omega = 1 \Longrightarrow s_{11} $es un valor finito entre 0 y 1  (Hay algo de reflexión pero no es total, comportamiento que se da en la frecuencia de corte)

Si $ \omega = \infty \Longrightarrow s_{11} = 1 $ (Hay total reflexión, el cuadripolo filtra altas frecuencias)

Analizando el parámetro $s_{21}$: $$ s_{21} = \frac{1}{2s^3+2s^2+2s+1} $$

Si $ \omega = 0 \Longrightarrow s_{21} = 1 $ (Hay total transmisión, el cuadripolo NO filtra bajas frecuencias)

Si $ \omega = 1 \Longrightarrow s_{21}$ es un valor finito entre 0 y 1 (Hay transmisión pero no es total, comportamiento que se da en la frecuencia de corte)

Si $ \omega = \infty \Longrightarrow s_{21} = 0 $ (No hay transmisión, el cuadripolo filtra altas frecuencias)

En base a estos comportamientos se puede afirmar que el cuadripolo analizado se comporta como un filtro pasa bajos.
