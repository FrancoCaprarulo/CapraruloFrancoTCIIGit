Me ayudo de una herramienta de Python para separar $ s_{11(s)} $ de $ s_{11(-s)} $

```python
import sympy as sp
#import numpy as np
import scipy.signal as sig
import splane as tc2
#import matplotlib as plt


s = sp.symbols('s ', complex=True)

s11sq = (-s**6+6*s**4-45*s**2)/(-s**6+6*s**4-45*s**2+225)

s11 = tc2.modsq2mod_s(s11sq).evalf(4)

tc2.display(s11)
```

Teniendo en cuenta que $s_{11}$ representa el coeficiente de reflexin en el puerto 1 del cuadripolo se puede decir que corresponde al de un filtro pasa bajos porque:

Si $ \omega = \infty $(Centro de la banda de stop) $\Longrightarrow s_{11} = 0 dB $ (Hay total reflexin, el cuadripolo filtra altas frecuencias)
