Necesitamos $ \alpha_{MAX} = 1dB $ como atenuación máxima en la banda de paso, entonces $ \epsilon = 10^{\frac{\alpha_{MAX}}{10}} - 1 $

$$ \epsilon^2 = 0.2589 $$

Y también queremos que: $ \alpha_{MIN} = 35dB $ (mínima atenuación en la banda de corte) y $ \alpha_{MIN} = 10log[1 + \epsilon^2(3.5)^{2n}] $ , para eso iteramos n hasta superar la cota inferior determinada por el $ \alpha_{MIN} $

Al llegar a $ n = 4 $ se obtiene una atenuación de $ 37.66dB $ en la banda de stop, suficiente para cumplir con los requisitos de la plantilla.

Primero se intenta desarrollar un filtro pasa bajos de esas características para luego hacer la transformación en frecuencia correspondiente. Por lo tanto para un filtro Butter de orden 4 los polos se encuentran en:

$ \theta_1 = \frac{5}{8} \pi \wedge \theta_2 = \frac{7}{8} \pi $ y por consecuencia $ Q_1 = 1.3066 \wedge Q_2 = 0.5412 $



Teniendo en cuenta las estructuras pasivas de segundo orden ya estudiadas y normalizando en frecuencia e impedancia $ L = Q \wedge C = \frac{1}{Q} $ calculamos los valores normalizados de los componentes...

$ L_1 = Q_1 = 1.3066 , L_2 = 0.5412 , C_1 = \frac{1}{Q_1} = 0.7653 \wedge C_2 = \frac{1}{Q_2} = 1.8477 $

Para desnormalizar consideramos las siguientes ecuaciones: $ L' = \frac{L\Omega}{\omega_B} \wedge C' = \frac{C}{\omega_B\Omega_Z} $

Ademas: $ \omega_B = \omega_P\epsilon^{\frac{1}{n}} $

Realizando el cálculo se obtiene que: $ \omega_B \cong 18573.31 \frac{rad}{s} $ y reemplazando los valores en las ecuaciones de L' y C' hallamos los valores de los componentes desnormalizados.

$$ L_1' = 70.35 mHy , L_2' = 29.14 mHy , C_1' = 41.2 nF \wedge C_2' = 99.48 nF $$
